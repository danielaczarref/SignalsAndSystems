```python
import math
import cmath 
import numpy
import matplotlib.pyplot as plt

def funcao (a, b, c, c1, c2):
    delta = b**2 - (4*a*c)
    #y = list(range(0,30))
    #x = list(range(0,30))
    #print(y)
    #print(x)
    tempo = numpy.arange(0, 30, 0.01)
    y=[2]
    if (delta > 0):
        print("delta maior que zero")
        r1 = (-b + math.sqrt(delta))/(2*a)
        r2 = (-b - math.sqrt(delta))/(2*a)
        #y = c1*math.e(r1*x) + c2*math.e(r2*x)
    elif (delta == 0):
        print("delta igual a zero")
        r1 = (-b)/(2*a)
        r2 = r1
        #y = c1*math.e(r1*x) + x*c2*math.e(r2*x)
    elif (delta < 0):
        print("delta menor que zero")
        r1 = (-b + cmath.sqrt(delta))/(2*a)
        r2 = (-b - cmath.sqrt(delta))/(2*a)
        lam = r1.real 
        mi = r2.imag
        #print("y(x) = e^("+lam+"x)* ("+c1+")*cos(")
        for item in range (1, len(tempo), 1):
            t = tempo[item]
            #print(t)
            y.append((math.exp(lam*t))*(c1*math.cos(mi*t) + c2*math.sin(mi*t)))
            #print(item)
            #print(y)
            #plt.plot(tempo, y)
        #print(y)
        plt.plot(tempo,y)
    return r1, r2

funcao(1, 0.125, 1, 2, c2)
```
